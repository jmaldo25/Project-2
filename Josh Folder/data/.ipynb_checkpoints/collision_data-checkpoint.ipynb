```python
# Dependencies and Set Up
import pandas as pd
import numpy as np

# CSV Read
collision_csv = "./vehicle_crashes_nonull.csv"

# Create DataFrame from csv
collision_data = pd.read_csv(collision_csv)
collision_data.head()
```

```python
# Create copy to do slicing without compromising original data
collision2 = collision_data.copy()
collision2.head()
```

```python
# Calculate totals of contributing factor - total
collision2.value_counts("contributing_factor_vehicle_1")
```

```python
# Calculate totale of vehicle type - total
collision2.value_counts("vehicle_type_code1")
```

```python
# Create groupby so that data can be counted by borough - Reason for crash
borough_gb = collision2.groupby("borough")

borough_collision_reason = borough_gb["contributing_factor_vehicle_1"].count()
borough_collision_reason
```

```python
# Create groupby so that data can be counted by borough - Type of vehicle involved
#borough_gb = collision2.groupby("borough")

borough_collision_vehicle = borough_gb["vehicle_type_code1"].count()
borough_collision_vehicle
```

```python
# Created groupby so that info is stored by borough and contributing factor
collision_reason_grp = collision2.groupby(["borough", "contributing_factor_vehicle_1"])

collision_count = collision_reason_grp["contributing_factor_vehicle_1"].count()
collision_count
```

```python
# Created dataframe of info sorted
collision_df = pd.DataFrame(collision_count)
collision_df
```

```python
# Renaming the colum so that index can be reset
new_collision_df = collision_df.rename(columns={"contributing_factor_vehicle_1": "contributing_factor_total"})
```

```python
# Reset index so that columns are created of the groupbys
borough_collision_df = new_collision_df.reset_index()
borough_collision_df
```

```python
# Starting on the vehicles involved in collisions
```

```python
# Created groupbyt for borough and vehicle type
collision_vehicle_grp = collision2.groupby(["borough", "vehicle_type_code1"])

vehicle_count = collision_vehicle_grp["vehicle_type_code1"].count()
vehicle_count
```

```python
# Created dataframe of info
vehicle_df = pd.DataFrame(vehicle_count)
vehicle_df
```

```python
# Renamed column so that columns can be used
new_vehicle_df = vehicle_df.rename(columns={"vehicle_type_code1": "vehicle_type_total"})
```

```python
new_vehicle_df
```

```python
# Reset index so that columns are created
borough_vehicle_df = new_vehicle_df.reset_index()
borough_vehicle_df
```

```python
# Created groupby for vehicle type
only_vehicle_grp = collision2.groupby("vehicle_type_code1")

only_vehicle_count = only_vehicle_grp["vehicle_type_code1"].count()
only_vehicle_count
```

```python
# Created dataframe of info
only_vehicle_df = pd.DataFrame(only_vehicle_count)
only_vehicle_df
```

```python
# Renamed column so that columns can be used
new_solovehicle_df = only_vehicle_df.rename(columns={"vehicle_type_code1": "vehicle_type_total"})
new_solovehicle_df
```

```python
# Reset index so that columns are created
clean_solovehicle_df = new_solovehicle_df.reset_index()
clean_solovehicle_df
```

```python
# Exported dataframe to csv
borough_collision_df.to_csv(r'borough_collision.csv', index=False)
```

```python
# Exported dataframe to csv
borough_vehicle_df.to_csv(r'borough_vehicle.csv', index=False)
```

```python
# Exported dataframe to json
borough_collision_df.to_json (r'borough_collision.json')
```

```python
# Exported dataframe to csv
clean_solovehicle_df.to_csv(r'all_vehicle_collision.csv', index=False)
```
